```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_cbs = pd.read_csv('../../data/raw/CBS_Kerncijfers.csv', sep=';')
df_cbs.describe()
```

```python
df_cbs = df_cbs[df_cbs["k_0Tot15Jaar_8"] != 0]
df_cbs = df_cbs[df_cbs["k_15Tot25Jaar_9"] != 0]
df_cbs = df_cbs[df_cbs["AantalInwoners_5"] != 0]

df_cbs = df_cbs[df_cbs["k_0Tot15Jaar_8"] != '']
df_cbs = df_cbs[df_cbs["k_15Tot25Jaar_9"] != '']
df_cbs = df_cbs[df_cbs["AantalInwoners_5"] != '']

df_cbs = df_cbs[df_cbs["k_0Tot15Jaar_8"].notna()]
df_cbs = df_cbs[df_cbs["k_15Tot25Jaar_9"].notna()]
df_cbs = df_cbs[df_cbs["AantalInwoners_5"].notna()]
df_cbs.describe()
```

```python
df_cbs.to_csv('../../data/processed/CBS_Kerncijfers_cleaned.csv', sep=';', index=False)
```

```python
df_groei = pd.read_csv('../../data/processed/df_v10_ratio_temp.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for wijk in df_groei["Wijk"].unique():
    plt.figure()
    
    subset1 = df_groei[df_groei["Wijk"] == wijk]
    
    ax = sns.lineplot(
        data=subset1,
        x="Jaar",
        y="k_0Tot15Jaar_8",
        marker="o"
    )
    
    ax.grid(True)
    plt.title(f"Populatie per jaar (0-15) - {wijk}")
    plt.xlabel("Jaar")
    plt.ylabel("Aantal inwoners")
    plt.show()
```

```python
from re import S


subset2 = df_groei

subset2['Peildatum'] = pd.to_datetime(subset2['Peildatum'])

subset2 = subset2[subset2['Jaar'].isin([2018, 2021])]

subset2 = subset2.pivot(index='Wijk', columns='Jaar', values='k_0Tot15Jaar_8')

subset2["groei"] = subset2[2021] - subset2[2018]

subset2["percentage_groei"] = (subset2["groei"] / subset2[2018]) * 100
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    data=subset2.reset_index(),
    x='Wijk',
    y='percentage_groei',
)

plt.title('percentage groei leeftijdsgroep 0-15 jaar per wijk (2018-2021)')

ax.axhline(0, color='black')
ax.set_ylim(subset2['percentage_groei'].min(), subset2['percentage_groei'].max())
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    data=subset2.reset_index(),
    x='Wijk',
    y='groei',
)

plt.title('groei leeftijdsgroep 0-15 jaar per wijk (2018-2021)')

ax.axhline(0, color='black')
ax.set_ylim(subset2['groei'].min(), subset2['groei'].max())
plt.show()
```
